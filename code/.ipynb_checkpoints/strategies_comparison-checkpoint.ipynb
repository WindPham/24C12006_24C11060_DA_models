```python
import numpy as np
import pandas as pd
import os
import sys
```

### Đọc dữ liệu

```python
df = pd.read_csv("../data/onlret_transaction.csv", dtype={"CustomerID": str})
```

```python
df.head(2)
```

### Tạo dữ liệu monthly index

```python
df["d"] = pd.to_datetime(df.InvoiceDate).dt.date
```

```python
df["m"] = pd.to_datetime(df.InvoiceDate).dt.to_period("M").dt.to_timestamp()
```

```python
df.head(2)
```

```python
strategy_01 = df.groupby("m").agg(revenue = ("TotalPrice", "sum"), 
                    cnt_cus = ("CustomerID", "nunique"),
                    cnt_pro = ("StockCode", "nunique")
                   ).reset_index()
strategy_02 = df.groupby(by=["m", "StockCode"]).agg(revenue = ("TotalPrice", "sum"), 
                    cnt_cus = ("CustomerID", "nunique"),
                    cnt_pro = ("Quantity", "sum")
                   ).reset_index()
strategy_03 = df.groupby(by=["m", "CustomerID"]).agg(revenue = ("TotalPrice", "sum"), 
                    cnt_cus = ("StockCode", "nunique"),
                    cnt_pro = ("Quantity", "sum")
                   ).reset_index()
strategy_04 = df.groupby("m").agg(revenue = ("TotalPrice", "sum"), 
                    cnt_cus = ("CustomerID", "nunique"),
                    cnt_pro = ("StockCode", "nunique"),
                    cnt = ("Quantity", "sum")      
                   ).reset_index()
```

```python
strategy_01.shape, strategy_01.shape, strategy_01.shape, strategy_04.shape
```

```python
strategy_01.to_csv("../data/strategy_01_monthly.csv", index=False)
strategy_02.to_csv("../data/strategy_02_monthly.csv", index=False)
strategy_03.to_csv("../data/strategy_03_monthly.csv", index=False)
strategy_04.to_csv("../data/strategy_04_monthly.csv", index=False)
```
