```python
import numpy as np
import pandas as pd
import os
import sys
```

### Đọc dữ liệu thô

```python
df = pd.read_csv("../data/onlret.csv")
```

```python
df.head(3)
```

```python
df["TotalPrice"] = df.Quantity * df.UnitPrice
```

```python
df[df.TotalPrice<0].head(3)
```

```python
df["Status"] = df.TotalPrice.apply(lambda x: 1 if x>=0 else 0)
```

```python
df.Status.value_counts()
```

### Xử lý, giảm bộ nhớ

```python
df.isna().sum()
```

```python
df[df.Description.isna()].CustomerID.value_counts(dropna=False)
```

```python
df["CustomerType"] = df.CustomerID.fillna("").apply(lambda x: 0 if x=="" else 1)
```

```python
df.head(2)
```

```python
df1=df[["StockCode", "Description"]].drop_duplicates()
df2=df.drop(columns="Description")
```

```python
df1.to_csv("../data/onlret_description.csv", index=False)
```
